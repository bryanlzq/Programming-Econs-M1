```r
require(MASS)
# Set the correlation parameter and mean
beta = 0.5
# variance co variance matrix
SIGMA = matrix(c(1,beta,beta,1), ncol=2) 
MU = c(2.0, 1.0)

# Set the sample size
N = 50

# Draw your sample
out <- mvrnorm(N, mu = MU, Sigma = SIGMA)
```

```r
# The data set is named `out`, which we split into y and X
y <- out[, 2]
# add a new column with vector of 1
X <- cbind(rep(1, N),out[, 1])
# generate matrix X that has additional column of 1s

# Now carry out intermediate calculations
XT = t(X)
XTX = XT%*%X
invXTX = solve(XTX)
XTy = XT%*%y
beta = invXTX %*% XTy

# Now add this line to the plot
plot(out)
abline(lm(out[,2]~out[,1]), col="red") # regression line (y~x) 
abline(a=beta[1], b=beta[2], col="blue")
```
