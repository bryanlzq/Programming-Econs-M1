```r
w0 <- 100 
p <- runif(1) 
T <- 10  

w <- rep(0, T+1)
x <- rep(0, T)

sprintf("Probability of winning %s", p)
sprintf("You started with $%s", w0)

w[1] = w0

for(t in 1:T){
    u <- runif(1)
    if(u <= 1 - p){
        print("You LOSE!")
        w[t+1] <- w[t] - 1
    }else{
        print("You WIN!")
        w[t+1] <- w[t] + 1
    }
    print(sprintf("Now you've got $%s.", w[t+1]))
}

time = 0:T
plot(time, w, "l")
```
