```r
#' The function contains two variables.
#'
#' @param a: value
#' @param b: value
#' @return summation of both variables
#' @export

my_add <- function (a, b){
    return(a + b)
}
my_add(1,1)
```

```r
#' The function contains a letter or word.
#'
#' @param pattern: a pattern to search for
#' @param word: a string in which to search
#' @return: The index to the first occurence of pattern in word.
#' @export

whereis <- function(pattern, word){
        return(grep(pattern, word))
}
whereis('I','Im lost')
```

```r
#' @rep(value, n) Repeat value for n number of times.
#' The function contains a variable
#'
#' @param x: a variable
#' @summ is define as 0
#' @for: shown the list of numbers repeated in x
#' @the numbers are added together.
#' @return: Tabulate the sum
#' @export

x <- rep(1, 10)
my_sum <- function(x){
    summ <- 0
    for(i in x){
        summ <- summ + i
    }
    return(summ)
}
my_sum(x)
```

```r
#' The function contains a number and a function
#'
#' @param n: number of occurance
#' @param F_inv: an inverse function
#' @return The F_inv of the random uniform sample
#' @export

rand_sample <- function(n, F_inv=function(x) x){
    unif_sample <- runif(n)
    return(F_inv(unif_sample))
}
rand_sample(5)              
```

```r
sample <- function(n, F_inv=function(x) x){
    sam <- runif(n,0,10)
    return(F_inv(sam))
}
mc_mean <- function(n, F_inv=function(x) x){
    x <- sample(n, F_inv)
    return(sum(x)/n)
}
```

```r
sims <- function(n_vec, F_inv=function(x) x){
    MCs <- rep(0, length(n_vec))
    i <- 1
    for(n in n_vec){
        MCs[i] <- mc_mean(n, F_inv)
        i <- i + 1
    }
    return(MCs)
}
N <- seq(0, 10000, 100)
MCs <- sims(N)
plot(N, MCs, "l")
```

```r
# Given pdf
# X ~ f(x) = 3x^2

# cdf is the integral of pdf
# cdf = F(x) = x^3

# The inverse of cdf
# F^-1(x) = x^(1/3)
```

```r
sample <- function(n, F_inv=function(x) x^(1/3))){
    sam <- runif(n)
    return(F_inv(sam))
}

mc_mean <- function(n, F_inv=function(x) x^(1/3))){
    x <- sample(n, F_inv)
    return(sum(x)/n)
}

mc_mean(5)
```
